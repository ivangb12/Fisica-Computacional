```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_csv('ramiro.txt',skiprows=4, sep='\s+')
```

```python
df0.head()
```

```python
df = pd.DataFrame(df0)
```

```python
df.dtypes
```

```python
df['FECHA']=pd.to_datetime(df.apply(lambda x: x['DD/MM/AAAA']+' '+x['HH:MM'],1),dayfirst=True)
df = df.drop(['DD/MM/AAAA','HH:MM'],1)
```

```python
df.head()
```

```python
df.describe()
```

```python
df_tmp = df[df.TEMP >24]
df_select = df_tmp[df_tmp.TEMP < 25]
df_select
```

```python
df_tmp = df[df.TEMP >14]
df_select = df_tmp[df_tmp.TEMP < 24]
df_select
```

```python
plt.figure(); df.VELR.plot(); plt.legend(loc='best')
plt.title("Velocidad rafagas")
plt.ylabel("rapidez (m/s)")
plt.grid(True)
plt.show()
```

```python
df1 = df[['TEMP','HR']]
plt.figure(); df1.plot(); plt.legend(loc='best')
plt.title("Variacin de la temperatura y humedad relativa")
plt.ylabel("Temp C/(%) HR")
plt.grid(True)
plt.show
```

```python
plt.plot_date(x=df.FECHA, y=df.TEMP, fmt="b-")
plt.title("Variacion de la Temperatura")
plt.ylabel("TempC")
plt.xlabel("Fecha")
plt.grid(True)
plt.show()
```

```python
df2 = df[['VELS','VELR']]
plt.plot_date(x=df.FECHA, y=df2, fmt="b-")
plt.title("Velocidad viento y rafagas")
plt.ylabel("Rapidez (m/s)")
plt.grid(True)
plt.show
```

```python
plt.plot_date(x=df.FECHA, y=df.DIRS, fmt="b-")
plt.title("Direccion de los vientos")
plt.ylabel("Grados")
plt.grid(True)
plt.show
```
